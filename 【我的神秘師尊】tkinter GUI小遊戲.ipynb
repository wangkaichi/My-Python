```python
#【我的神秘師尊】tkinter GUI小遊戲
#使用基本text、button、listbox與showinfo

import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

#點擊後回傳資料
def do_it():
    print("乖徒平身，此後跟著為師修練，必有所成！")
    neverend(win)

def printstory1():
    lb=tk.Label(text="你揹著背包在國度裡旅行著，每天都很開心。\
    \n有一天，你看見有一個人昏迷在路上，你要救他嗎？", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    helpp(win)
    btn1=tk.Button(text="「救他」")
    btn1.config(bg="#ffffff", fg="#080808", width=15, height=1, command=printstory2)
    btn1.pack()

def printstory2():
    lb=tk.Label(text="你檢查過後，發現他是因為飢餓口渴才昏迷在地，\
    \n於是打開自己的背包，想看看有沒有東西可以救他。", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    btn2=tk.Button(text="打開背包")
    btn2.config(bg="#ffffff", fg="#080808", width=17, height=1, command=backpack1)
    btn2.pack()

def backpack1():
    listbox1 = tk.Listbox(win, height=4, bg="#fadcdc", fg="#1515b3")
    listbox1.insert(tk.END, "檯燈", "果汁機", "藍芽鍵盤", "礦泉水")
    listbox1.pack()
    tk.Button(win, text="關閉背包", bg="#ffffff", fg="#080808", width=17, command=printstory3).pack()

def printstory3():
    lb=tk.Label(text="「這都是些什麼東西？」你不明白自己背包裡怎麼會有這些奇怪的東西。\
    \n還好有一罐礦泉水，於是你從背包裡取出，準備要餵給他喝時，聽見他對你說道。\
    \n「我、我要喝蘋果汁…………」", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    apple(win)
    btn3=tk.Button(text="「製作蘋果汁給他喝」")
    btn3.config(bg="#ffffff", fg="#080808", width=18, height=1, command=printstory4)
    btn3.pack()

def printstory4():
    lb=tk.Label(text="要製作蘋果汁就必須要有蘋果，\
    \n於是你往旁邊的森林走，終於找到了一棵蘋果樹，樹下掉落幾顆新鮮的蘋果。", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5",fg="#543236")
    lb.pack()
    btn4=tk.Button(text="獲得蘋果")
    btn4.config(bg="#ffffff", fg="#080808", width=20, height=1, command=backpack2)
    btn4.pack()

def backpack2():
    listbox2 = tk.Listbox(win, height=5, bg="#fadcdc", fg="#1515b3")
    listbox2.insert(tk.END, "檯燈", "果汁機", "藍芽鍵盤", "礦泉水", "蘋果")
    listbox2.pack()
    tk.Button(win, text="取出果汁機、礦泉水、蘋果", bg="#ffffff", fg="#080808", width=20, command=backpack3).pack()

def backpack3():
    listbox3 = tk.Listbox(win, height=2, bg="#fadcdc", fg="#1515b3")
    listbox3.insert(tk.END, "檯燈", "藍芽鍵盤")
    listbox3.pack()
    juice(win)
    tk.Button(win, text="給他喝下蘋果汁", bg="#ffffff", fg="#080808", width=20, command=printstory5).pack()

def printstory5():
    lb=tk.Label(text="喝下蘋果汁後，他重新獲得精神，激動地抓住你的手，對你說：\
    \n「你是我的救命恩人，我將會把我的畢生絕學傳授給你！」", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    teach(win)
    btn5=tk.Button(text="「徒弟參見師尊！」")
    btn5.config(bg="#ffffff", fg="#080808", width=19, height=1, command=do_it)
    btn5.pack()

#使用showinfo消息提示框
class helpp(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "「不想管他，不救」"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("must", "必須要救")

class apple(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "「要求太多了吧，不理」"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("hahahaha", "必須給他喝蘋果汁")

class juice(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "用果汁機將蘋果打成蘋果汁"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("get", "獲得蘋果汁")

class teach(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "「遇到瘋子，轉身就走」"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("surprise", "不能走，必須接受")

class neverend(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "<未完待續>"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("don't click", "就跟你說待續，不要按了！")


#設定
win=tk.Tk()
win.title("Faraway Kingdom")
win.geometry("600x830")
win.resizable(False,False)
win.iconbitmap("castle.ico")
win.config(bg="#b7e6f7")
win.attributes("-topmost",True)
win.attributes("-alpha",1.0)

#story start
lb=tk.Label(text="在很久很久以前，有一個美麗的國度，\
\n國度裡的每個人都相敬如賓，生活過得非常幸福。")
lb.config(bg="#fafac5",fg="#543236",font=("微軟正黑體",11))
lb.pack()
btn0=tk.Button(text="進入國度")
btn0.config(bg="#ffffff", fg="#080808", width=17, height=1, command=printstory1)
btn0.pack()

#執行
win.mainloop()
```
