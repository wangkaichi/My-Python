```python
#【我的神秘師尊-前傳】tkinter GUI小遊戲
#使用基本text、button、listbox與showinfo

import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

#點擊後回傳資料
def do_it():
    print("待緣分足矣，總有一天，乖徒會見到為師的！")
    neverend(dow)

#刪除全部text
def delete_all():
    text.delete(1.0, "end")

def printstory1():
    lb=tk.Label(text="你前往信中所說的老地方，但沒看到人，只看見地上有一個綠色的包裹。", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    btn1=tk.Button(text="打開綠色包裹")
    btn1.config(bg="#ffffff", fg="#080808", width=17, height=1, command=backpack1)
    btn1.pack()

#使用listbox
def backpack1():
    listbox1 = tk.Listbox(dow, height=4, bg="#02DF82", fg="#000000")
    listbox1.insert(tk.END, "女僕裝", "貓耳朵", "貓尾巴", "白絲襪")
    listbox1.pack()
    wear(dow)
    tk.Button(dow, text="「穿上包裹中的衣服配件」", bg="#ffffff", fg="#080808", width=20, command=printstory2).pack()

def printstory2():
    lb=tk.Label(text="包裹中有一封信，信中寫道要你變裝潛入女僕咖啡廳執行任務。\
    \n「原來啊…………」你鬆了一口氣，還好不是要做一些奇奇怪怪的事情。", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    btn2=tk.Button(text="潛入女僕咖啡廳")
    btn2.config(bg="#ffffff", fg="#080808", width=20, height=1, command=printstory3)
    btn2.pack()

def printstory3():
    lb=tk.Label(text="潛入女僕咖啡廳後，跟著線索，你在第四個門後的第三個抽屜裡找到第二個藍色包裹。", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5",fg="#543236")
    lb.pack()
    btn3=tk.Button(text="打開藍色包裹")
    btn3.config(bg="#ffffff", fg="#080808", width=20, height=1, command=backpack2)
    btn3.pack()

def backpack2():
    listbox2 = tk.Listbox(dow, height=1, bg="#2894FF", fg="#000000")
    listbox2.insert(tk.END, "一封信")
    listbox2.pack()
    tk.Button(dow, text="打開信件", bg="#ffffff", fg="#080808", width=20, command=backpack3).pack()

def backpack3():
    text=tk.Text(dow, width=20, height=5, bg="#fadcdc")
    text.config(padx=5, pady=5, font=("微軟正黑體",10))
    text.insert(tk.INSERT, "這是一封信，無須銷毀。\n")
    text.insert(tk.INSERT, "能找到這裡，你很聰明，\n")
    text.insert(tk.INSERT, "恭喜你通過考驗，\n")
    text.insert(tk.INSERT, "有成為我徒弟的資格！\n")
    text.insert(tk.END, "by.你的師尊")
    text.pack()
    destroy(dow)
    btn4=tk.Button(dow, text="師尊你在哪裡", command=printstory1)
    btn4.config(bg="#ffffff", fg="#080808", width=11, height=1, command=printstory4)
    btn4.pack()

def printstory4():
    lb=tk.Label(text="「總有一天你會見到我的。」聲音從遠方傳來，你未見其人。\
    \n想來你的師尊定是武功高強、法力無邊！", font=("微軟正黑體", 11))
    lb.config(bg="#fafac5", fg="#543236")
    lb.pack()
    btn5=tk.Button(text="「弟子靜候師尊！」")
    btn5.config(bg="#ffffff", fg="#080808", width=19, height=1, command=do_it)
    btn5.pack()

#使用showinfo消息提示框
class wear(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "「裝作沒看見，關上包裹」"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("must", "必須要穿")

class destroy(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "閱 後 即 焚"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("Understand?", "就跟你說不用銷毀，聽不懂膩？")

class neverend(ttk.Frame):
    def __init__(self, hint):
        ttk.Frame.__init__(self, hint)
        self.pack()
        self.button = ttk.Button(self)
        self.button["text"] = "<未完待續>"
        self.button["command"] = self.hintkey
        self.button.pack()
    def hintkey(self):
        showinfo("don't click", "就跟你說待續，不要按了！\
        \n快去脫下女僕裝吧哈哈哈哈哈！")


#設定
dow=tk.Tk()
dow.title("Letter")
dow.geometry("600x830")
dow.resizable(False,False)
dow.iconbitmap("castle.ico")
dow.config(bg="#b7e6f7")
dow.attributes("-topmost",True)
dow.attributes("-alpha",1.0)

#story start
lb=tk.Label(text="有一天你收到一封神秘的信。")
lb.config(bg="#fafac5",fg="#543236",font=("微軟正黑體",11))
lb.pack()

text=tk.Text(dow, width=20, height=8, bg="#fadcdc")
text.config(padx=5, pady=5, font=("微軟正黑體",10))
text.insert(tk.INSERT, "這是一封信。\n")
text.insert(tk.INSERT, "千萬不要跟第二個人說起。\n")
text.insert(tk.INSERT, "你看到信之後，\n")
text.insert(tk.INSERT, "立刻前往老地方，\n")
text.insert(tk.INSERT, "有重要的任務要交給你，\n")
text.insert(tk.INSERT, "讀完信後立刻焚毀，\n")
text.insert(tk.INSERT, "記住了嗎？\n")
text.insert(tk.END, "by.神秘人")
text.pack()
   
btn0=tk.Button(text="閱後即焚")
btn0.config(bg="#ffffff", fg="#080808", width=17, height=1, command=delete_all)
btn0.pack()
btn1=tk.Button(dow, text="前往老地方", command=printstory1)
btn1.config(bg="#ffffff", fg="#080808", width=17, height=1, command=printstory1)
btn1.pack()

#執行
dow.mainloop()
```
